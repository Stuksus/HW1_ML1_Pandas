#### 7. [0.5 балла] Найдите курсы по выбору, на которые записывались как студенты 18-го года набора, так и студенты 17-го года.

Проверим версию о том, что в fall_1 и spring_1 представлено все множество возможных курсов на выбор для осени и весны соответственно, а не одно из подмножеств

```python
def get_full_set(frame,column1,column2,column3):
    """
    Функция собирающая 3 списка в множество
    Входящие параметры:
    - frame: Данные
    - column1: Первый столбец 
    - column2: Второй столбец
    - column3: Третий столбец
    Выход:
    - Множество, составленное из всех элементов 3-х списков
    """
    subject_list = []
    subject_list += frame[column1].unique().tolist()
    subject_list += frame[column2].unique().tolist()
    subject_list += frame[column3].unique().tolist()
    return set(subject_list)

# Создадим полное множество из всех осенних предметов
subject_list_fall = get_full_set(frame,'fall_1','fall_2','fall_3')
# Создадим полное множество из всех весенних предметов
subject_list_spring = get_full_set(frame,'spring_1','spring_2','spring_3')
```

```python
def differences_in_sets(full_set,frame,column1,column2,column3):
    """
    Функция, вычисляющая наличие различий в общем 
    и частных множествах
    Входящие параметры:
    - full_set: Множество возможных значений
    - frame: Данные
    - column1: Первый столбец 
    - column2: Второй столбец
    - column3: Третий столбец
    Выход:
    Если данные множества явялются подмножествами 
    full_set:
    - Элементы, которые не попали в подмножество
    - Выводит 1
    В противном случае:
    - Пустой словарь
    - Выводит 0
    """
    diff = {column1:[],column2:[],column3:[]}
    diff[column1] += full_set - set(frame[column1].unique().tolist())
    diff[column2] += full_set - set(frame[column2].unique().tolist())
    diff[column3] += full_set - set(frame[column3].unique().tolist())
    
    if (len(diff[column1]) == 0 and len(diff[column2]) == 0 and
    len(diff[column3]) == 0):
        return diff,0
    else:
        return diff,1
```

```python
differences_in_sets(subject_list_spring,frame,'spring_1','spring_2','spring_3')
```

Алгоритм распределения студентов по курсам:
1. По умолчанию на каждой дисциплине по выбору у 3 и 4 курсов может учиться 1 группа (до 30 студентов). Исключения описаны ниже. На blended-дисциплинах для четверокурсников количество мест не ограничено.
2. Проводится первая волна отбора. Для каждой дисциплины формируется список тех, кто указал её первым приоритетом (если студент должен выбрать два курса по выбору, то для него дисциплины, которые он указал первым и вторым приоритетом, рассматриваются как дисциплины первого приоритета). 4. Если студент не попал на необходимое количество курсов по итогам трёх волн, с ним связывается учебный офис и решает вопрос в индивидуальном порядке.

```python
def del_students(group,last_place):
    """
    Учитывает количесвто оставшихся мест на предмет
    и в случае, если произошел перенабор студентов,
    то удалет излишнее количество с конца
    Входные параметры:
    - group: группа
    - last_place: данные о оставшихся местах по предметам
    Выходные данные:
    - Если количество сводных мест больше или равно 0:
        - Возвращает полученную группу без изменений
    - Если количество сводных мест меньше 0(произошел перенабор):
        - Возвращает группу, за исключением определенного кол-ва
        (на которое произошел перенабор) студентов с конца
    """
    # Получим название предмета и по нему получим число мест
    count_place_to_save = last_place.loc[group.index[0][0]]
    # В зависимости от полученного значения вернем группу
    if count_place_to_save >= 0:
        return group
    else:
        return group.iloc[:int(count_place_to_save)]
```

```python
# Обработаем данные на наличие перенабора
full_del_last = (full_fall1_fall2_group
                 .groupby('subject', as_index=False)
                 .apply(lambda x: del_students(x,last_place)))
```
